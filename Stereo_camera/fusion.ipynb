```python
import csv

def weighted_average(dist1, dist2, threshold):
    if dist1 < threshold:
        weight1 = 0.4
        weight2 = 0.6
    else:
        weight1 = 0.6
        weight2 = 0.4
    
    return (dist1 * weight1 + dist2 * weight2) / (weight1 + weight2)

data1 = []
with open('added_trackerpoints1.txt', 'r') as file:
    next(file)
    for line in file:
        frame, distance = map(float, line.strip().split(','))
        data1.append(distance)

data2 = []
with open('tracker_points3.txt', 'r') as file:
    next(file)
    for line in file:
        frame, distance = map(float, line.strip().split(','))
        data2.append(distance)

output_file = 'fusion_distances1.csv'
with open(output_file, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Index', 'Weighted_Average_Distance'])
    
    for i, (dist1, dist2) in enumerate(zip(data1, data2), start=1):
        weighted_avg = weighted_average(dist1, dist2, 4)
        writer.writerow([i, weighted_avg])
```

```python
import csv

def weighted_average(angle1, angle2, frame, low_threshold, high_threshold):
    if frame < low_threshold or frame > high_threshold:
        weight1 = 0.7
        weight2 = 0.3
    else:
        weight1 = 0.7
        weight2 = 0.3
    
    return (angle1 * weight1 + -angle2 * weight2) / (weight1 + weight2)

data1 = []
with open('added_angle1.txt', 'r') as file:
    next(file)
    for line in file:
        frame, angle = map(float, line.strip().split(','))
        data1.append((frame, angle))

data2 = []
with open('angle_tracker3.txt', 'r') as file:
    next(file)
    for line in file:
        frame, angle = map(float, line.strip().split(','))
        data2.append((frame, angle))

output_file = 'fusion_angles1.csv'
with open(output_file, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Index', 'Weighted_Average_Angle'])
    
    for i, ((frame1, angle1), (frame2, angle2)) in enumerate(zip(data1, data2), start=1):
        weighted_avg = weighted_average(angle1, angle2, frame1, 200, 400)
        writer.writerow([i, weighted_avg])

```

```python
import csv

def weighted_average(dist1, dist2, dist3, threshold):
    if dist1 < threshold:
        weights = [0.7, 0.2, 0.1]
    else:
        weights = [0.7, 0.2, 0.1]
    
    return (dist1 * weights[0] + dist2 * weights[1] + dist3 * weights[2]) / sum(weights)

def read_distances(filename):
    distances = []
    with open(filename, 'r') as file:
        next(file)
        for line in file:
            frame, distance = map(float, line.strip().split(','))
            distances.append(distance)
    return distances

# Read data from three files
data1 = read_distances('added_trackerpoints1.txt')
data2 = read_distances('tracker_points3.txt')
data3 = read_distances('lidar_points.txt')

output_file = 'fusion_distances1.csv'
with open(output_file, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Index', 'Weighted_Average_Distance'])
    
    for i, (dist1, dist2, dist3) in enumerate(zip(data1, data2, data3), start=1):
        weighted_avg = weighted_average(dist1, dist2, dist3, 4)
        writer.writerow([i, weighted_avg])
```
